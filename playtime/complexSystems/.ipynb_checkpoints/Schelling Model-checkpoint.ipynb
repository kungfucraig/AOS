```python
import random

EMPTY = None
SURFER = 'U'
SWIMMER = 'W'
MAX_ADJ = 8
P = 0.333

class Board:
    def __init__(self, size, nsurfers, nswimmers):
        self.size = size
        self.cells = [EMPTY for i in range(self.size*self.size)]
        assert(nswimmers+nsurfers <= size*size/2)
        for i in range(nsurfers):
            self.addAgent(SURFER)
        for i in range(nswimmers):
            self.addAgent(SWIMMER)
            
    def getIndex(self, x, y):
        return x*self.size+y
    
    def setCell(self, x, y, agentType):
        assert(not self.getAgent(x, y))
        assert(agentType in [EMPTY, SURFER, SWIMMER])
        self.cells[self.getIndex(x,y)] = agentType
        
    def addAgent(self, agentType):
        added = False
        while not added:
            x = random.randrange(self.size)
            y = random.randrange(self.size)
            if not self.getAgent(x, y):
                added = True
                self.setCell(x, y, agentType)
    
    def moveAgent(self, x, y):
        index = self.getIndex(x, y)
        a = self.cells[index]
        if a:
            self.cells[index] = EMPTY
            self.addAgent(a)
                
    def getAgent(self, x, y):
        return self.cells[self.getIndex(x,y)]
                
    def getAdjacent(self, x, y):
        index = self.getIndex(x, y)
        baseIndexes = [-1, 1, -self.size-1, -self.size+1, self.size+1, self.size-1]
        indexes = [(index+i)%(self.size*self.size) for i in baseIndexes]
        return [self.cells[i] for i in indexes if self.cells[i]]
        
        
    def echo(self):
        print "*"*10
        for i in range(self.size):
            l = ""
            for j in range(self.size):
                cell = self.cells[self.getIndex(i,j)]
                l += cell if cell else '-'
                l += ' '
                 
            print l 
        
class Simulator:
    
    def __init__(self, board, max_iterations):
        self.board = board
        self.max_iterations = max_iterations
        
    def run(self):
        for i in range(self.max_iterations):
            #b.echo()
            if self.update():
                break;
        print "Iterations=",i
              
    # return true if nothing moved
    def update(self):
        agentsToMove = []
        for i in range(self.board.size):
            for j in range(self.board.size):
                a = self.board.getAgent(i, j)
                if a:
                    adj = self.board.getAdjacent(i, j)
                    same = sum([x==a for x in adj])
                    nadj = len(adj)
                    if nadj and float(same)/float(nadj) < P:
                        agentsToMove.append((i, j))

        for a in agentsToMove:
            self.board.moveAgent(a[0], a[1])
        return len(agentsToMove) == 0
                
                
  
DIM = 12
SURFERS = int(1.0/3.0*DIM*DIM/2.0)
SWIMMERS = int(2.0/3.0*DIM*DIM/2.0)
ITERS = 10
b = Board(DIM, SURFERS, SWIMMERS)
b.echo()
s = Simulator(b, ITERS)
s.run()
b.echo()
```
