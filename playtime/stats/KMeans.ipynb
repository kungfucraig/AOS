```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# cultural model http://jasss.soc.surrey.ac.uk/12/1/6/appendixB/Axelrod1997.html
```

```python
# Create fake data
n = 1000
s1 = np.random.multivariate_normal([0, 0], [[2, 0], [0, 3]], n)
s2 = np.random.multivariate_normal([-4, 2], [[5, 0], [0, 1]], n)
s = np.concatenate((s1, s2))

# Assign a color representing the distribution each point was taken from
c = ["green" for i in range(n)]
c.extend(["blue" for i in range(n)])

# Scatter Plot
plt.scatter(s[:,0], s[:, 1], c=c)
```

```python
NUM_CLUSTERS = 5
NUM_ITERATIONS = 50
TOLERANCE = 0.01
CENTROID_COLOR = "red"
COLOR_MAP = ("green", "blue", "cyan", "magenta", "yellow", "red")

class KMeans:
    def __init__(self, max_iterations=50, min_error=0.01):
        self.iterations_remaining = max_iterations
        self.min_error = 0.01

    def __call__(self, data, initial_centroids):
        """
        Args:
            data a N x DIM array where each row is considered to be a sample point of dimension DIM
            centroids a K x DIM array where each row is the centroid of a cluster of dimension DIM
        """
        n = data.shape[0]
        dim = data.shape[1]
        assert(initial_centroids.shape[1] == dim)
        k = initial_centroids.shape[0]
    
        # kxn array where the row represents the distances of each point in data
        # from the centroid
        def differences(x):
            diff = np.asmatrix(initial_centroids - x)
            m = diff * diff.transpose()
            out = np.sqrt(np.diag(m))
            return out
        distances = np.apply_along_axis(differences, axis=1, arr=data)
    
        # for each sample figure out what cluster it belongs to
        cluster_count = np.zeros((k, 1))
        new_centroids = np.zeros((k, dim))
        cluster_assignment = np.zeros((n, 1), dtype=np.int8)
        for i in range(n):
            data_point = data[i, :]
            data_point_distances = distances[i,:]
            assigned_cluster = np.argmin(data_point_distances)
        
            cluster_assignment[i, 0] = assigned_cluster
            new_centroids[assigned_cluster, :] += data[i, :]
            cluster_count[assigned_cluster, 0] += 1
        
        new_centroids = new_centroids / cluster_count
        centroid_diff = new_centroids - initial_centroids
        error = np.linalg.norm(centroid_diff)
        if self.iterations_remaining <= 0 or error < self.min_error:
            return (new_centroids, cluster_assignment)
        else:
            self.iterations_remaining -= 1
            return self(data, new_centroids)
    
    
selector = np.random.randint(s.shape[0],size=NUM_CLUSTERS)
centroids = s[selector,:]

kmeans = KMeans()
(final_centroids, cluster_assignments) = kmeans(s, centroids)

# Assign a color representing the distribution each point was taken from
d = [COLOR_MAP[x] for x in cluster_assignments[:, 0]]
d.extend([CENTROID_COLOR for i in range(NUM_CLUSTERS)])

areas = [3*np.pi for x in cluster_assignments[:, 0]]
areas.extend([12*np.pi for i in range(NUM_CLUSTERS)])


# Concatenate all data
t = np.concatenate((s, final_centroids))


# Scatter Plot
plt.scatter(t[:,0], t[:, 1], s=areas, c=d)
```
