```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rng = 1
omega = 1/2.0
x = np.linspace(-rng, rng, 100)
xx = 1j * x * 2 * np.pi * omega
y = np.exp(xx)

f, (ax1, ax2) = plt.subplots(2, sharex=True)
ax1.plot(x, np.real(y))
ax2.plot(x, np.imag(y))

plt.show()
```
